```python
from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)
```

```python
import pandas as pd
```

```python
pd.concat([iris.data, iris.target], axis=1).plot.scatter(
    x='petal length (cm)', y='petal width (cm)', c='target',
    colormap='viridis', figsize=(10,4)
)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
print(iris.target.unique())
```

```python
from sklearn.linear_model import Perceptron
```

# Exercise 1

```python
perceptron0 = Perceptron(random_state=42)
y_train_0 = (y_train == 0)
y_test_0 = (y_test == 0)
perceptron0.fit(X_train, y_train_0)

y_pred_train_0 = perceptron0.predict(X_train)
acc_train_0 = perceptron0.score(X_train, y_train_0)

y_pred_test_0 = perceptron0.predict(X_test)
acc_test_0 = perceptron0.score(X_test, y_test_0)

# wght0 = (perceptron.intercept_, perceptron.coef_[0][0], perceptron.coef_[0][1])
wght0 = (perceptron0.coef_[0][0], perceptron0.coef_[0][1], perceptron0.coef_[0][2])
```

```python
perceptron1 = Perceptron(random_state=42)
y_train_1 = (y_train == 1)
y_test_1 = (y_test == 1)
perceptron1.fit(X_train, y_train_1)

y_pred_train_1 = perceptron1.predict(X_train)
acc_train_1 = perceptron1.score(X_train, y_train_1)

y_pred_test_1 = perceptron1.predict(X_test)
acc_test_1 = perceptron1.score(X_test, y_test_1)

# wght1 = (perceptron.intercept_, perceptron.coef_[0][0], perceptron.coef_[0][1])
wght1 = (perceptron1.coef_[0][0], perceptron1.coef_[0][1], perceptron1.coef_[0][2])
```

```python
perceptron2 = Perceptron(random_state=42)
y_train_2 = (y_train == 2)
y_test_2=(y_test==2)
perceptron2.fit(X_train, y_train_2)

y_pred_train_2 = perceptron2.predict(X_train)
acc_train_2 = perceptron2.score(X_train, y_train_2)

y_pred_test_2 = perceptron2.predict(X_test)
acc_test_2 = perceptron2.score(X_test, y_test_2)

wght2 = (perceptron2.coef_[0][0], perceptron2.coef_[0][1], perceptron2.coef_[0][2])
```

```python
per_acc=[(acc_train_0, acc_test_0), (acc_train_1, acc_test_1), (acc_train_2,acc_test_2)]
per_wght=[wght0,wght1,wght2]
print(per_acc)
print(per_wght)
```

```python
import pickle as pkl
with open("per_acc.pkl", "wb") as f:
    pkl.dump(per_acc, f)
```

```python
with open("per_wght.pkl", "wb") as file:
    pkl.dump(per_wght, file)
```

```python
from sklearn.neural_network import MLPClassifier
import numpy as np

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

model = MLPClassifier (activation='relu' , solver='sgd', max_iter=1000, learning_rate_init=0.1)
model.fit(X, y)

print("accuracy: ", model.score(X, y))
print("predict: ", model.predict(X))
print("hidden-layer-coef: ", model.coefs_[0])
print("enter-layer-coef: ", model.coefs_[1])
print("hidden-layer-bias: ", model.intercepts_[0])
print("enter-layer-bias: ", model.intercepts_[1])
```

```python
with open('mlp_xor_fixed.pkl', 'wb') as f:
    pkl.dump(model,f)
```
